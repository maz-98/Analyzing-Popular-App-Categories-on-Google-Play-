lets starts by building a dictionary

```python
# Group By 'App' and get the max review for each group
reviews_max = android_df.groupby('App')['Reviews'].max()
reviews_max["Instagram"]
```

```python
reviews_max
```

```python
# create an empty List to store elemend data
android_clean = []
# create an empty List to keep track pf already added apps
already_added = []
# Iterate through each row the DataFrame
for index, row in android_df.iterrows():
    name = row["App"]
    n_reviews = row["Reviews"]
    
    # check if the currect app has the maximum number of reiews and has not been added before
    if (reviews_max[name]) == n_reviews and (name not in already_added ):
        android_clean.append(row)
        already_added.append(name)









```

```python
android_clean = pd.DataFrame(android_clean)
```

# Removing English Alphabets


```python
def is_english(app_name):
    lst = []
    for i in app_name:
        if ord(i) > 127 :
            lst.append(False) 
        else:  
            lst.append(True)
    non_ascii = 0 
    for j in lst :
        if j == False:
            non_ascii += 1
            
    if non_ascii > 3:
        return False
    else:
        return True
```

```python
is_english("my ")
```

```python
android_english = android_clean["App"].apply(is_english)
```

```python
android_english = android_clean[android_clean["App"].apply(is_english)]
```
